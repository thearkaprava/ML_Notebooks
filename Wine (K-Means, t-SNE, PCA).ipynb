```python
import pandas as pd
import numpy as np
import chardet
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 

pd.read_csv('Wine.csv', encoding = 'Latin-1')
```

```python
with open("Wine.csv", 'rb') as f:
    result = chardet.detect(f.read()) 

print(result['encoding'])
df_raw = pd.read_csv('Wine.csv', encoding=result['encoding'])
df = df_raw.loc[:,'14.23':'1065']
labels = df_raw.loc[:,'1']
df.head()
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=1)
clusters = kmeans.fit_predict(df)
kmeans.cluster_centers_.shape
```

```python
from scipy.stats import mode

pred_labels = np.zeros_like(clusters)
for i in range(3):
    mask = (clusters == i)
    pred_labels[mask] = mode(labels[mask])[0]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(labels, pred_labels)
```

# t-SNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, init='random', random_state=0)
data_proj = tsne.fit_transform(df)

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_proj)

pred_labels = np.zeros_like(clusters)
for i in range(3):
    mask = (clusters == i)
    pred_labels[mask] = mode(labels[mask])[0]

accuracy_score(labels, pred_labels)
```

**We see that the accuracy score is marginally greater when we reduce the number of components to 2 using t-SNE**

```python
import matplotlib
from sklearn.manifold import TSNE
import seaborn as sns

sns.set_style("darkgrid", {"axes.facecolor": ".95"})
%config InlineBackend.figure_format = 'svg'
```

```python
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
proj = pd.DataFrame(data_proj)
proj.columns = ["comp_1", "comp_2"]
proj["pred_labels"] = pred_labels
sns.lmplot("comp_1", "comp_2", hue = "pred_labels", data = proj ,fit_reg=False)
print("ready")
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,pred_labels)
```

**The silhouette score indicates moderately good clustering.**

# PCA

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

data_scaled = scaler.fit_transform(df)

from sklearn.decomposition import PCA

"""
    pca can also be initialized as pca = pca(0.95) : fraction of variance explained
"""


from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principal_components = pca.fit_transform(data_scaled)
                    
data_proj = pd.DataFrame(data = principal_components, 
                           columns = ['pc_1', 'pc_2'])

kmeans = KMeans(n_clusters=10, random_state=42)
clusters = kmeans.fit_predict(data_proj)

# Permute the labels
pred_labels_pca = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    pred_labels_pca[mask] = mode(labels[mask])[0]

# Compute the accuracy
accuracy_score(labels, pred_labels_pca)
```

**The accuracy score increases drastically when we use PCS for dimensionality reduction.**

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,pred_labels_pca)
```

**However the silhouette score is almost 0 indicating a lot of border cases.**
